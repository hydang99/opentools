# OpenTools Tool Demo (no agents)

This notebook shows how to run individual tools (e.g. `Wolfram_Math_Tool`, `Search_Engine_Tool`, `Wiki_Search_Tool`) directly, without going through an agent or `UnifiedSolver`. Moreover, we also describe how to test a tool with provided test file (e.g. `Calculator_Tool`)


## 1. Tool Running

```python
import json
import sys

# Make `opentools` importable when running from `docs/demo/`
sys.path.insert(0, "..")
sys.path.insert(0, "src")

from opentools.tools.wolfram_math.tool import Wolfram_Math_Tool
from opentools.tools.search_engine.tool import Search_Engine_Tool
from opentools.tools.wiki_search.tool import Wiki_Search_Tool
from opentools.tools.calculator.tool import Calculator_Tool

def show(label, result):
    """Pretty-print a tool result dict."""
    print(f"=== {label} ===")
    if isinstance(result, dict):
        print(json.dumps(result, indent=2, default=str))
    else:
        print(result)
```

### a. Wolfram_Math_Tool

Direct call into Wolfram Alpha for a simple derivative/integral query.


```python
wolfram = Wolfram_Math_Tool()
wolfram_result = wolfram.run(query="derivative of x^2 + 3*x at x = 5")
show("Wolfram_Math_Tool", wolfram_result)
```

### b. Search_Engine_Tool

Web search using the Google Custom Search API wrapper.


```python
search = Search_Engine_Tool()
search_result = search.run(
    query="OctoTools agentic framework complex reasoning",
    num_results=3,
    output_format="markdown",
)
show("Search_Engine_Tool", search_result)

show("Search_Engine_Tool", search_result["result"])
```

### c. Wiki_Search_Tool

Wikipedia search for a concept related to the demos.


```python
wiki = Wiki_Search_Tool()
wiki_result = wiki.run(
    operation="search",
    query="OctoTools (agentic framework)",
    limit=3,
    output_format="markdown",
)
show("Wiki_Search_Tool", wiki_result)

show("Wiki_Search_Tool", wiki_result["result"])
```
